# Breast Cancer Classification model using Sklearn and breast_cancer Dataset from sklearn

## importing the data

```python
from sklearn.datasets import load_breast_cancer
b_cancer = load_breast_cancer()
b_cancer
```

```python
# converting the b_cancer data into pandas DataSet
import pandas as pd
cancer_data = pd.DataFrame(b_cancer['data'],columns=b_cancer['feature_names'])
cancer_data
```

```python
# joining the target column to our cancer dataframe
# the target contains the value 0 or 1 
# 0 : doesnt has breast cancer 
# 1 : has breast cancer
cancer_data['target'] = b_cancer['target']
cancer_data
```

```python
#check if there is any missing datafields in our dataframe
cancer_data.isna().sum()
```

**all values are 0 that means there is no datafield missing**

```python
# split the data into X : features and Y : labels (target)
X = cancer_data.drop('target',axis=1)
Y=cancer_data['target']
```

```python
X.head()
```

```python
Y.head()
```

# ML model (RandomForestClassifier) 

**now we are going to use sklearn's classification model RandomForestClassifier**
**it is an ensemble model which means it uses many different smaller models for classification** <br>
**you can learn more about model here: https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html**<br>
**learn more about ensemble here: https://scikit-learn.org/stable/modules/ensemble.html**

```python
#import the model from sklearn.ensemble
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=200)
# n-estimators defines the number of smaller models to use 
```

```python
# now we have to split the data into train and test sets because we cant use same data to train and test the model
# we already have a method to split the dataset (X and Y) into train and test sets 
from sklearn.model_selection import train_test_split
```

**learn about sklearn.model_selection.train_test_split here :https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html**

```python
import numpy as np
```

```python
np.random.seed(42)
x_train , x_test , y_train , y_test = train_test_split(X,Y,test_size=0.2)
# test_size represent the proportion of the dataset to include in the test split (e.g 0.2 means 80% train data and 20% test data) 
# we used np.random.seed(42) so as if we run the cell multiple time it still gives the same split
```

```python
# now we fit/train the model on x_train and y_train
model.fit(x_train,y_train)
```

```python
# now our model is ready to predict 
y_preds = model.predict(x_test)
y_preds[:5]
```

```python
np.array(y_test[:5])
```

**y_preds is an array which contains the predicted values and y_test is the ground truth (true value)**

# Evaluation

**Score method**<br>

 model.score() methods takes the test split of our X(feature) and Y(target) dataset and evaluates our model and gives us Coefficient of Determination (R^2) read here:https://www.investopedia.com/terms/c/coefficient-of-determination.asp

```python
model.score(x_test,y_test)
```

**Cross-validation**<br>

read here :https://scikit-learn.org/stable/modules/cross_validation.html <br>

cross_val_score takes model , feature dataset , target dataset as parameter
<br>
**cv=5 means how many subsets should be out dataset should be split and trained and tested upon**

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X,Y,cv=5)
```

**sklearn.metrics.accuracy_score**<br>
read here : https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html#sklearn.metrics.accuracy_score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_preds)
```

**ROC (Receiver operating characteristic) curve**<br>
read here: https://scikit-learn.org/stable/modules/generated/sklearn.metrics.roc_curve.html <br>
roc_curve takes ground truth and Target scores probability estimates of the positive class  as parameter and returns False Positive Rate(fpr) , True Positive rate(tpr) and Thresholds

```python
y_probs = model.predict_proba(x_test) #predicts probability of the class
y_probs_1 = y_probs[:,1]
```

```python
from sklearn.metrics import roc_curve
fpr , tpr , thrd = roc_curve(y_test,y_probs_1)
```

```python
## now we want to plot this curve into a graph for better visualization therefore we'll create a plotting function
import matplotlib.pyplot as plt

def plot_roc(fpr,tpr):
    """
    plots roc curve figure given false positive rate(fpr) and true positive rate (tpr)
    """
    #plot roc
    plt.plot(fpr,tpr,color='orange',label='ROC')
    #plot line with no predective power 'BaseLine'
    plt.plot([0,1],[0,1],color='darkblue',label='Guessing',linestyle='--')
    
    #customize plot
    plt.xlabel('False Positive Rate(fpr)')
    plt.ylabel('True Positive Rate(tpr)')
    plt.title('Receiver Operating Charateristic(ROC)')
    plt.legend()
    plt.show()
```

```python
## now we plot our roc graph usinf plot_roc function
plot_roc(fpr,tpr)
```

```python
## now to show how a perfect roc curve looks like 
fpr , tpr , thrd = roc_curve(y_test,y_test)
plot_roc(fpr,tpr)
```

**Confusion matrix**
<br>A confusion matrix is a summary of prediction results on a classification problem. The number of correct and incorrect predictions are summarized with count values and broken down by each class.<br>
read here : https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html

```python
from sklearn.metrics import confusion_matrix

# it takes parameter ground truths and predictions
confusion_matrix(y_test,y_preds)
```

```python
# for better visualization
pd.crosstab(y_test,y_preds,rownames=['Actual'],colnames=['predicted'])
```

```python
#for better visualization we gonna use sklearn's ConfusionMatrixDisplay
from sklearn.metrics import ConfusionMatrixDisplay
# from estimators method takes model , feature dataset and label/target dataset as parameters
ConfusionMatrixDisplay.from_estimator(model,X,Y)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test,y_preds) # from predcition takes ground truth and target dataset (test) as parameter
```

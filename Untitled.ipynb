## importing the data

```python
from sklearn.datasets import load_breast_cancer
b_cancer = load_breast_cancer()
b_cancer
```

```python
# converting the b_cancer data into pandas DataSet
import pandas as pd
cancer_data = pd.DataFrame(b_cancer['data'],columns=b_cancer['feature_names'])
cancer_data
```

```python
# joining the target column to our cancer dataframe
# the target contains the value 0 or 1 
# 0 : doesnt has breast cancer 
# 1 : has breast cancer
cancer_data['target'] = b_cancer['target']
cancer_data
```
